```python
race_results = pd.read_csv('dataset/Race_Results.csv')
```

```python
race_results.head()
```

```python
race_results.info()
```

```python
# Replace '\\N' with NaN
race_results['position'].replace('\\N', np.nan, inplace=True)

# Change the 'position' column type to int
race_results['position'] = race_results['position'].astype(float).astype('Int64')
```

```python
race_results.info()
```

```python
race_results['position'].describe()
```

```python
driver_details = pd.read_csv('dataset/Driver_Details.csv')
```

```python
driver_details['driver_name'] = driver_details['forename'] + " " + driver_details['surname']
```

```python
driver_details.head()
```

```python
race_schedule['year'].value_counts()
```

```python
team_details = pd.read_csv('dataset/Team_Details.csv')
```

```python
team_details.head()
```

```python
team_details.rename(columns={'name': 'team_name'}, inplace=True)
```

```python
team_details.info()
```

```python
# df = pd.merge(sprint_race_results, driver_details, on='driverId')
df = pd.merge(race_results, race_schedule, on='raceId')
```

```python
df['year'].describe()
```

```python
# df = pd.merge(df, race_schedule, on='raceId')
df = pd.merge(df, driver_details, on='driverId')
```

```python
df['year'].describe()
```

```python
df = pd.merge(df, team_details, on='constructorId')
```

```python
df.head()
```

```python
df.info()
```

```python
df[['year', 'position', 'driver_name', 'team_name']].head()
```

```python
top_3_positions = df[df['position'] <= 3]

top_3_positions = top_3_positions[[
    'year', 'position', 'driver_name', 'team_name']].sort_values(by=['position', 'year'])

top_3_positions.head(10) 
```
